## MODEL BUILDING

## 1.POPULARITY BASED RECOMMENDATION SYSTEM

```python
Ratings_with_name=Ratings.merge(Books,on='ISBN')
```

```python
Ratings_with_name
```

```python
#First we have grouped the book-title from book df and ratings coloumn from Ratings df
num_rating_df=Ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
#it gives us book title and no of ratings on the book
```

```python
#we have calculated the avg rating of each book
avg_rating_df=Ratings_with_name.groupby('Book-Title').mean(numeric_only=True)['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head(50)
```

```python
popular_df
```

```python
popular_df=popular_df.merge(Books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Year-Of-Publication','num_ratings','avg_ratings','Image-URL-M']]
```

```python
#we have  built our popularity based recommendation system
popular_df
```

## 2.COLLABRATIVE FILTERING BASED RECOMMENDATION SYSTEM

```python
x=Ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
actually_intrested=x[x].index
```

```python
#Here only 92000/200000 users have rated any book
```

```python
#it filters out and give us the rows that have users that are actually intrested and has rated over 200 books
filtered_rating=Ratings_with_name[Ratings_with_name['User-ID'].isin(actually_intrested)]
```

```python
#filtering on basis of books
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
#706books*810users
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
similarity_score
```

```python
def recommend(book_name):
    # Fetch the index of the book
    indices = np.where(pt.index == book_name)[0]
    if len(indices) == 0:
        print(f"Book '{book_name}' not found in the dataset.")
        return
    
    index = indices[0]
    
    # Find similar items
    similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x: x[1], reverse=True)[1:6]
    data= []
    # Print similar items
    for i in similar_items:
        item=[]
        temp_df=(Books[Books['Book-Title']==pt.index[i[0]]])
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
        
        return data
```

```python
recommend('1984')
```
